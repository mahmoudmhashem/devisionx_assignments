```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flower = cv2.imread("flower.jpeg")
flower = cv2.cvtColor(flower, cv2.COLOR_BGR2RGB)
```

```python
imshow(flower)
```

```python
filter_size = 15

filter = np.ones((filter_size, filter_size)) / (filter_size**2)

flower_blured_maually = cv2.filter2D(flower, -1, filter)
flower_blured_opencv = cv2.blur(flower, (filter_size, filter_size))

flower_blured_gaussian = cv2.GaussianBlur(flower, (filter_size, filter_size), 0)
flower_blured_median = cv2.medianBlur(flower, filter_size)
```

```python
fig, axes = plt.subplots(3, 2)
fig.figure.set_size_inches(10, 15)

axes[0, 0].imshow(flower)
axes[0, 0].set_title("flower_original")

axes[1, 0].imshow(flower_blured_maually)
axes[1, 0].set_title("flower_blured_maually")

axes[1, 1].imshow(flower_blured_opencv)
axes[1, 1].set_title("flower_blured_opencv")

axes[2, 0].imshow(flower_blured_gaussian)
axes[2, 0].set_title("flower_blured_gaussian")

axes[2, 1].imshow(flower_blured_median)
axes[2, 1].set_title("flower_blured_median")

same = np.all(flower_blured_maually == flower_blured_opencv)
if same : print("bluered image manually is the same as image bluered by opencv")
else : print("bluered image manually is not the same as image bluered by opencv")

plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
fig.figure.set_size_inches(10, 15)

axes[0, 0].imshow(thankyou)
axes[0, 0].set_title("thankyou")

axes[0, 1].imshow(thankyou_eroded)
axes[0, 1].set_title("thankyou_eroded")

axes[1, 0].imshow(thankyou_dilated)
axes[1, 0].set_title("thankyou_dilated")

axes[1, 1].imshow(~thankyou_diff, cmap='Greys')
axes[1, 1].set_title("thankyou_diff")

plt.show()
```
